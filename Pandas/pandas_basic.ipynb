```python
import pandas as pd
dict1={'key1':[1,2,3,4],
      'key2':[3,4,5,7]} 
data1 = pd.DataFrame(dict1)
dict2={'key3':[7,4,1,5],
      'key4':[2,0,6,4]} 
data2 = pd.DataFrame(dict2)
data1.merge(data2,how='cross')
```

```python
import pandas as pd
dict1={'key1':[1,2,3,4],
      'key2':[3,4,5,7]} 
data1 = pd.DataFrame(dict1)
dict2={'key1':[2,4,1,5],
      'key4':[2,0,6,4]} 
data2 = pd.DataFrame(dict2)
data1.merge(data2,how='cross')
data1.merge(data2,how='right') #one col_name must be common
data1.merge(data2,how='left')
data1.merge(data2,how='inner') 
```

```python
import pandas as pd
dict1={'key1':[1,2,3,4],
      'key2':[3,4,5,7]}
data1 = pd.DataFrame(dict1,index=[1,2,3,4])
dict2={'key3':[2,4,1,5],
      'key4':[2,0,6,4]}
data2 = pd.DataFrame(dict2,  index=[1,2,5,6])
data1.join(data2) #by default how='left'
data1.join(data2,how='outer')
data1.join(data2,how='inner')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sb.load_dataset('iris')

# Create a scatter plot
sb.scatterplot(x='sepal_length', y='sepal_width', data=iris)

# Show the plot
plt.show()
```

```python
sb.load_dataset("tips")
```

```python
tips=sb.load_dataset("tips")
sb.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
#count smoker and non-smoker
tips['smoker'].value_counts()
```

```python
sb.relplot(x="total_bill",y='tip',data=tips,hue='smoker')
```

```python
sb.relplot(x="total_bill",y='tip',data=tips,style='size')
```

```python
sb.relplot(x="total_bill",y='tip',data=tips,style='size',hue='time')
```

```python
sb.catplot(x="day",y='total_bill',data=tips)
```

```python
sb.catplot(x="smoker",y='total_bill',data=tips)
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips) #distributionof data + histogram(counting the freq,quantity)
```

```python
sb.pairplot(tips)
```
